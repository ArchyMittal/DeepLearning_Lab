Experiment 2: Creating and Printing Tensors

```python
#SCALAR TENSOR
a=tf.constant(5)
print(a)

#VECTOR TENSOR
v=tf.constant([1,2,3,4,])
print(v)

#MATRIX TENSOR
m=tf.constant([[1,2],[3,4]])
print(m)

t3=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print(t3)
```

Experiment 3: Basic Tensor Operations

```python
# --------SCALAR---------

a=tf.constant(10)
b=tf.constant(20)

# ADDITION
print(a+b)

#SUBTRACTION
print(a-b)

#MULTIPLICATION
print(a*b)

#DIVISION
print(a/b)
```

```python
# --------VECTOR---------

a=tf.constant([10,20,30])
b=tf.constant([40,50,60])

# ADDITION
print(a+b)

#SUBTRACTION
print(a-b)

#MULTIPLICATION
print(a*b)

#DIVISION
print(a/b)
```

```python
# --------2D-MATRIX---------

a=tf.constant([[10,20],[30,40]])
b=tf.constant([[50,60],[70,80]])

# ADDITION
print(a+b)

#SUBTRACTION
print(a-b)

#MULTIPLICATION
print(a*b)

#DIVISION
print(a/b)
```

Experiment 4: Data Types and Shape of Tensors

```python
# -------Scalar---------
a = tf.constant(10)

print("Value:", a)
print("Data Type:", a.dtype)
print("Shape:", a.shape)
print("Number of Dimensions:", a.ndim)
print("\n")


# -------Vector---------
b = tf.constant([1, 2, 3, 4])

print("Value:\n", b)
print("Data Type:", b.dtype)
print("Shape:", b.shape)
print("Number of Dimensions:", b.ndim)
print("\n")


# ----------2D Matrix------------
c = tf.constant([[1, 2, 3],
                      [4, 5, 6]])

print("Value:\n", c)
print("Data Type:", c.dtype)
print("Shape:", c.shape)
print("Number of Dimensions:", c.ndim)
```

Experiment 5: Converting NumPy Arrays to Tensors

```python
import numpy as np

numpy_array = np.array([[10, 20, 30],
                        [40, 50, 60]])

tensor = tf.convert_to_tensor(numpy_array)

print("NumPy Array:\n", numpy_array)
print("\nTensorFlow Tensor:\n", tensor)
```
